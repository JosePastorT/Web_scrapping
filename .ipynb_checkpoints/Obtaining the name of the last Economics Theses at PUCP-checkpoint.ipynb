```python
from selenium import webdriver                          # pip install selenium in anaconda prompt
# from selenium.webdriver.common.keys import Keys       # give access to the enter key and the skip key

from selenium.webdriver.common.by import By
# from selenium.webdriver.support.ui import WebDriverWait
# from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
import time          # waitting seconds  

import warnings
warnings.filterwarnings("ignore")
```

```python
driver = webdriver.Chrome("C:\Program Files (x86)\chromedriver.exe")     # each time Chrome updates, we have to update the executable in Program Files
driver.get("https://tesis.pucp.edu.pe/repositorio/handle/20.500.12404/1026")

# printing title of webpage
print("Título de la página web: ", driver.title)

# waiting for the page to load properly
time.sleep(1)


tesis_names1 = []

# finding common element
titles1 = driver.find_elements(By.CLASS_NAME, 'col-sm-9')

for title1 in titles1:
    tesis_name1 = title1.find_element(By.TAG_NAME, 'a').text
    lista1 = {
        'Título de la tesis': tesis_name1
        }
    
    tesis_names1.append(lista1)

# accessing "View more" section
search = driver.find_element(By.XPATH, '//*[@id="aspect_discovery_CollectionRecentSubmissions_p_recent-submission-view-more"]/a')
search.click()

# repeating this part of the code to access more tesis names

for i in range(5):
    search = driver.find_element(By.XPATH, '//*[@id="aspect_discovery_recentSubmissions_RecentSubmissionTransformer_div_main-recent-submissions"]/div[3]/ul/li[2]/a')
    search.click()
    
    # finding common element
    titles1 = driver.find_elements(By.CLASS_NAME, 'col-sm-9')
    
    for title1 in titles1:
        tesis_name1 = title1.find_element(By.TAG_NAME, 'a').text
        lista1 = {
            'Título de la tesis': tesis_name1
            }
        
        tesis_names1.append(lista1)
```

```python
df = pd.DataFrame(tesis_names1)


# exporting dataframe
df.to_excel('D:/1_PUCP/Ciclo 2022-2/Seminario de Investigación/Tesis anteriores/tesis_names.xlsx', header=True, index=False, encoding='utf-8')


driver.quit()
```
